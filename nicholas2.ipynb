```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('Alinity Dataset 1.csv')
df['datetimestamplocal'] = pd.to_datetime(df['datetimestamplocal'], format="%m/%d/%Y %H:%M")

# Extract only the time (HH:MM in 24-hour format) for the x-axis
df['time_of_day'] = df['datetimestamplocal'].dt.strftime('%H:%M')
```

```python
first_reagent = df[df['reagentmasterlotnumber'] == '63055F900']
first_reagent
first_reagent = first_reagent.sort_values(by='datetimestamplocal')
```

```python
for assaynumber, group in first_reagent.groupby('assaynumber'):
    plt.figure(figsize=(12, 6))
    plt.plot(group['time_of_day'], group['correctedcount'], marker='o', linestyle='-', label=f"assaynumber = {assaynumber}")

    # Improve x-axis formatting
    plt.xticks(rotation=45, ha='right')

    # Labels and title
    plt.xlabel("Time of Day (24-hour format)")
    plt.ylabel("Count")
    plt.title(f"Time of Day vs. Count (assaynumber = {assaynumber})")

    # Show the grid and legend
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()
```
